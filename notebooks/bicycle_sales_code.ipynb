```python
import pandas as pd
import numpy as np
df_calendar = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Calendar_Lookup.csv")
df_customer = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Customer_Lookup.csv",encoding='latin-1')
df_product = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Product_Lookup.csv")
df_product_categories = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Product_Categories_Lookup.csv")
df_product_subCategories = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Product_Subcategories_Lookup.csv")
df_returns = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Returns_Data.csv")
df_territories = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Territory_Lookup.csv")
df_sales_2020 = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Sales_Data_2020.csv")
df_sales_2021 = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Sales_Data_2021.csv")
df_sales_2022 = pd.read_csv(r"C:\Users\smash\OneDrive\Desktop\CDF\Bicycle store\data\raw\Sales_Data_2022.csv")

```

```python
def data_cleaning(df, name):
    """ stndardizing column names and removing duplicate rows"""
    print(f"Cleaning {name} dataset and has {len(df)} rows")
    df = df.drop_duplicates()
    print(f"{name} has {len(df)} rows after deletion")
    # Remove completely empty rows
    df = df.dropna(how='all')
    print(f"Missing values : ",df.isnull().sum())
    print("\n")
    return df

df_calendar = data_cleaning(df_calendar,"Calendar")
df_customer = data_cleaning(df_customer,"Customer")
df_product = data_cleaning(df_product,"Product")
df_product_categories = data_cleaning(df_product_categories,"Product Category")
df_product_subCategories = data_cleaning(df_product_subCategories,"Product SubCategories")
df_returns = data_cleaning(df_returns,"Returns")
df_sales_2020 = data_cleaning(df_sales_2020,"Sales 2020")
df_sales_2021 = data_cleaning(df_sales_2021,"sales 2021")
df_sales_2022 = data_cleaning(df_sales_2022,"Sales 2022")
df_territories = data_cleaning(df_territories,"Territories")
```

```python
#cleaning customet dataset
df_customer['AnnualIncome'] = df_customer['AnnualIncome'].fillna(df_customer['AnnualIncome'].median())
df_customer = df_customer[df_customer['FirstName'].notna()]
#cleaning product dataset
df_product = df_product[df_product['ProductKey'].notna()]
df_product = df_product[df_product['ProductCost']>0]
df_product = df_product[df_product['ProductPrice']>0]
#cleaning product category dataset
df_product_categories = df_product_categories[df_product_categories['ProductCategoryKey'].notna()]
#cleaning product subcategory dataset
df_product_subCategories = df_product_subCategories[df_product_subCategories['ProductSubcategoryKey'].notna()]
#cleaning returns dataset
df_returns = df_returns[df_returns['ReturnQuantity']>0]
```
